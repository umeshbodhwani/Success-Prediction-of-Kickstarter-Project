```python
import urllib
from urllib.request import urlopen
from bs4 import BeautifulSoup
import json
import pandas as pd
import time
```

```python
KICKSTARTER_SEARCH_URL = "https://www.kickstarter.com/discover/advanced?sort=newest&seed=2502593&page={}"
DATA_FILE = "kickstarter.csv"
projects_dump = []

page_start = 1
while True:
    url = KICKSTARTER_SEARCH_URL.format(page_start)
    print(url)   
    try:
        response = urlopen(url)
        html = response.read()
    except urllib.error.HTTPError as err:
        if err.code == 404:
            break
        elif err.code == 429:
            print("Sleeping for 5 mins......")
            time.sleep(300)
            response = urlopen(url)
            html = response.read()
            
    soup = BeautifulSoup(html, 'html.parser')
    project_details_divs = soup.findAll('div', {"class":"js-react-proj-card"})
   
    if len(project_details_divs) == 0:
        break
   
    for div in project_details_divs:
        project = json.loads(div['data-project'])
        projects_dump.append(project)
    page_start +=1
```

```python
Project_DB = pd.DataFrame()
for i in projects_dump:
    Project_DB = pd.concat([Project_DB,pd.DataFrame([i])])
```

```python
Project_DB.to_pickle("Project_DB.p")
```
